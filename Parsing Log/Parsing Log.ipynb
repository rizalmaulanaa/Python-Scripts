```python
import os
from dateutil.parser import parse
import pandas as pd
import numpy as np
import glob
```

```python
def Date(index):
    try:
        return parse(text[index])
    except ValueError:
        return False

def find_date(index):
    status = Date(index)
    while(status is False):
        index-=1
        status = Date(index)
    return str(status),text[index]

def file_name(name):
    fn = os.path.basename(name)
    source_application = fn.split('_')
    return fn,source_application[1]

dt = pd.read_excel('contoh transformasi.xlsx',sheet_name = 1)
```

```python
nama_file = glob.glob("*.log")
datas = []

for files in nama_file:
    data = open(files,'r')
    text = data.read().split('\n')
    data.close()
    # print(text)

    index = 0
    fn, source_application = file_name(files)

    for i in text:
        if 'Starting ORACLE instance' in i:
            event_date,raw = find_date(index)
            datas.append([fn,source_application,'os','alertdb',source_application,event_date,raw,None,i,None,'Start',None])
        elif 'Instance shutdown complete' in i:
            event_date,raw = find_date(index)
            datas.append([fn,source_application,'os','alertdb',source_application,event_date,raw,None,i,None,'Shutdown',None])
        index+=1

df = pd.DataFrame(np.array(datas),columns = dt.columns)
df
```

```python
writer = pd.ExcelWriter('Transformasi.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save()
```
