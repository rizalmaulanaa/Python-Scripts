```python
import numpy as np
import pandas as pd
import re

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop = open('stopword tala.txt','r')
stoplist = stop.read().split('\n')
stop.close()
path = '/Users/Rizal Maulana/Downloads/semester 7/TextMin/'
```

# Read Data

```python
data = pd.read_excel(path+'Dataset Kelompok 4.xlsx')
korpus = data['Komentar']
label = data['Hasil akhir']
korpus
```

# Text Preprocessing

```python
doc_term = []
for doc in korpus:
    token = re.findall(r'\b[A-Za-z]{1,}',doc)
    term = []
    for word in token:
        if word not in stoplist:
            term.append(stemmer.stem(word.lower()))
    doc_term.append(term)
    
doc_term
```

# Pembobotan Term

```python
term = np.unique(sum(doc_term,[]))
columns = [i for i in range(len(doc_term))]
dtm = pd.DataFrame(0,index=term,columns=columns)

for i,doc in enumerate(doc_term):
    for term in doc:
        dtm.loc[term,i] += 1
        
dtm
```

# Naive Bayes

```python
class Naive_Bayes(object):
    def __init__(self,df):
        self.df = df
    
    def train(self,df_train):
        self.prior = {j:(i/len(df_train)) for i,j in zip(df_train.value_counts(),df_train.value_counts().index)}
        train_group = df_train.groupby(df_train)
        for i in self.prior.keys():
            self.df[i] = 0

        for i in self.prior.keys():
            for j in self.df.index:
                atas = self.df.loc[j,train_group.get_group(i).index].sum()+1
                bawah = self.df[train_group.get_group(i).index].sum().sum()+len(df_train.index)
                self.df[i] = atas/bawah
        
        print('Training data finish')

    def test(self,df_test):
        hasil = {}
        for i in df_test.index:
            temp = {}
            for j in self.prior.keys():
#                 print(i,j,np.prod(self.df[self.df[i]>0][j])*self.prior[j])
                temp[j] = np.prod(self.df[self.df[i]>0][j])*self.prior[j]
            hasil[i] = max(temp,key=temp.get)
        
        return pd.Series(hasil)
    
    def akurasi(self,y_true,y_predict):
        total = 0
        for i in y_true.index:
            if y_true[i] == y_predict[i]:
                total += 1
                
        return total/len(y_true)
```

# K-Fold

```python
k_fold = []
k = 3
flag = [0,15]

for i in range(0,5):
    temp = []
    for j in flag:
        temp += list(label[j:j+k].index)
    k_fold.append(temp)
    flag = [h+k for h in flag]
    
k_fold
```

# Evaluasi

```python
rata = []
for k,i in enumerate(k_fold):
    df_train = label[~label.index.isin(i)]
    df_test = label[i]

    nb = Naive_Bayes(dtm)
    nb.train(df_train)
    y_predict = nb.test(df_test)
    akurasi = nb.akurasi(df_test,y_predict)
    rata.append(akurasi)
    print(y_predict)
    print('Data Uji ke-{} dengan akurasi {}'.format(k+1,akurasi))
    
print('\nRata Rata Akurasi {}'.format(np.mean(rata)))
```

```python
from sklearn.naive_bayes import MultinomialNB
dtm_c = dtm.copy().transpose()
for i in k_fold:
    train_target = label[~label.index.isin(i)]
    train_dtm = dtm_c.iloc[train_target.index]
    test_target = label[i]
    test_dtm = dtm_c.iloc[i]

    clf = MultinomialNB().fit(train_dtm,train_target)
    print(clf.predict(test_dtm))
```
